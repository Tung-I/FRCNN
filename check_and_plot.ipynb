```python
import json
import os
import re
from pathlib import Path
import numpy as np
import cv2
import copy
import random
import xml.etree.ElementTree as ET
from PIL import Image
from matplotlib import pyplot as plt
from tqdm import tqdm_notebook as tqdm
from pycocotools.coco import COCO
```

```python
def plot_box(im, boxes, coco=False, cls=None):
    for i in range(boxes.shape[0]):
        box = boxes[i]
        if cls is not None:
            if box[4] != cls:
                continue
        if coco:
            cv2.rectangle(im, (box[0], box[1]), (box[0]+box[2], box[1]+box[3]), (20, 255, 20), 2)
        else:
            cv2.rectangle(im, (box[0], box[1]), (box[2], box[3]), (20, 255, 20), 2)
    return im
```

```python
im_dir = '/home/tony/YCB_simulation/query/images'
nd_dir = '/home/tony/YCB_simulation/query/ndarray'
n = str(259).zfill(6)
boxes = np.load(os.path.join(nd_dir, n+'.npy'), allow_pickle=True)
im = cv2.imread(os.path.join(im_dir, n+'.jpg'))
plot_box(im, boxes, cls=2)
plt.imshow(im[:, :, ::-1])
```

```python
coco_json_path = '/home/tony/datasets/YCB2D/annotations/instances_pseudo2.json'
im_id = 1112
with open(coco_json_path, 'r') as f:
    data = json.load(f)
im_dir = '/home/tony/YCB_simulation/query/images'
im = cv2.imread(os.path.join(im_dir, str(im_id).zfill(6)+'.jpg'))
boxes = []
for ann in data['annotations']:
    if ann['image_id'] == im_id:
        boxes.append(ann['bbox']+[ann['category_id']])
boxes = np.asarray(boxes)
plot_box(im, boxes, coco=True)
plt.imshow(im[:, :, ::-1])
```
